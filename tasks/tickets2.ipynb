
## Завдання на залік з дисципліни ІТ ІАЗ.

### Білет 2

Створити прототип інформаційно-аналітичної системи для відображення інформації 
щодо втрат противника у артилерії за останні 11 днів 
та прогнозу на наступні 7 днів.
на основі наданого шаблону.

1. Створити базу даних exam_74, створити таблицю та заповнити її даними за допомогою sql-файлів з шаблону.
2. Backend на основі Python FastAPI (налаштувати параметри підключення до бази даних, та прописати оператор SELECT для вибірки необхідних полів: 
   date, location, [поле відповідно до завдання] as data_field)
3. Frontend на основі:
   - HTML(додати місця розташування відповідних візуальних елементів)
   - CSS (налаштувати кольорову схему)
   - JavaScript (встановити базовий колір графіку та задати координати маркерів)

4. Кольорова схема має відповідати основному кольору Олива відповідно до [брендбуку ЗСУ](https://sprotyvg7.com.ua/wp-content/uploads/2024/06/1_%D0%9A%D0%95%D0%A0-%D0%97-%D0%92%D0%98%D0%9A-%D0%93%D0%A0-%D0%A1%D0%A2%D0%98%D0%9B%D0%AE-%D0%A3-%D0%92%D0%86%D0%97-%D0%9A%D0%9E%D0%9C-%D0%97%D0%A1%D0%A3.pdf).
5. Координати маркерів на інтерактивній карті:
     A: [50.9258, 34.7242]
     B: [48.4134, 35.0280]
     C: [46.6208, 32.6556] 
6. Додатково натренувати модель машинного навчання прогнозування втрат противника у артилерії.
7. Відобразити на веб-сторінці прогноз на наступні 7 днів.
8. Запустити backend (uvicorn main:app --reload) та frontend (live server) частини інформаційно-аналітичної системи.
9. Надати роботу на перевірку.
